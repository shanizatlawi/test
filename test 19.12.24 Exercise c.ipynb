```python
import pandas as pd

# Loading the data 
orders_df = pd.read_csv('C:/Users/97254/TEST_CSV/orders.csv')
order_item_df = pd.read_csv('C:/Users/97254/TEST_CSV/order_item.csv')
items_df = pd.read_csv('C:/Users/97254/TEST_CSV/items.csv')
customers_df = pd.read_csv('C:/Users/97254/TEST_CSV/customers.csv')
```

```python
#5
# Function to print details for each table
def print_table_details(df, table_name):
    print(f"Details for table: {table_name}")
    
    # 1. Column names, data types, and count of non-null values
    print("\nColumn Names, Data Types, and Non-Null Count:")
    print(df.dtypes)
    print(df.count())  # Count of non-null values
    
    # 2. Summary Statistics: mean, min, max, median, etc.
    print("\nSummary Statistics (mean, min, max, median, etc.):")
    print(df.describe())
    
    # Calculate Median (50th percentile) for numeric columns only
    print("\nMedian (50th percentile) for numeric columns:")
    numeric_cols = df.select_dtypes(include=['number']).columns
    print(df[numeric_cols].median())
    
    # 3. Print the first 5 rows
    print("\nFirst 5 rows:")
    print(df.head())
    
    print("\n" + "-"*50 + "\n")

# Print the details for each table
print_table_details(orders_df, "Orders")
print_table_details(order_item_df, "Order_Item")
print_table_details(items_df, "Items")
print_table_details(customers_df, "Customers")
```
